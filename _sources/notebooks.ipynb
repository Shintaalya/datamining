**2. Membuat Statistik Deskriptif dari data (setiap atribut yang bertipe numerik)**

**MEDIAN**

```python
med = {'Apps':data['Apps'].median(),
		'Accept': data['Accept'].median(),'Enroll':data['Enroll'].median(),
    'Top10perc':data['Top10perc'].median(),'Top25perc':data['Top25perc'].median(),
    'F.Undergrad':data['F.Undergrad'].median(),'P.Undergrad':data['P.Undergrad'].median(),
    'Outstate':data['Outstate'].median(),'Room.Board':data['Room.Board'].median(),
    'Books':data['Books'].median(),'Personal':data['Personal'].median(),
    'PhD':data['PhD'].median(),'Terminal':data['Terminal'].median(),
     'S.F.Ratio':data['S.F.Ratio'].median(),
     'perc.alumni':data['perc.alumni'].median(),
     'Expend':data['Expend'].median(),
     'Grad.Rate':data['Grad.Rate'].median()
    }

df = pd.DataFrame(med, index=['median'])
df
```

```python
data["Apps"].median()
```

**3. SKEWED / KEMIRINGAN**

```python
sns.distplot(data['Apps'],kde=True,
             bins= int(150/5),color='red',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':4}).set(title='Positive Skewed')
  
  
```

```python
sns.distplot(data['Accept'],kde=True,
             bins= int(150/5),color='yellow',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':4}).set(title='Positive Skewed')
  
```

```python
sns.distplot(data['Terminal'],kde=True,
             bins= int(150/5),color='green',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':4}).set(title='Positive Skewed')
  
```

```python
sns.distplot(data['Enroll'],kde=True,
             bins= int(150/5),color='darkblue',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':4}).set(title='Positive Skewed')
  
```

```python
sns.distplot(data['Grad.Rate'],kde=True,
             bins= int(150/5),color='orange',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth':4}).set(title='Positive Skewed')
  
```
